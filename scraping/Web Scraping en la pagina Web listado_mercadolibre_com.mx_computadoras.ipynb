```python
import numpy as np
import pandas as pd
import requests
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
```

<h2 style="color:magenta">2.- Enviar petición HTTP para extraer Código Fuente de la Página Web</h2>

```python
pagina_web = requests.get('https://listado.mercadolibre.com.mx/computadoras')

print("\n","Si despliega <Response [200]> (entonces tu petición fue escuchada y aceptada)  : ",pagina_web)
print()

soup = BeautifulSoup(pagina_web.text,'html.parser')

print("\n",soup.prettify()) 
```

<h2 style="color:magenta">Vamos a extraer los títulos de las máquinas de esa Código Fuente de la Página Web</h2>


##### <span style="color:blue">Igualamos a (nombre_maquina) y realizamos una extracción de los títulos de las máquinas de esa Código Fuente de la Página Web y desplegamos y los colocamos todos los elementos encontrados en un arreglo como una lista.</span>

```python
#//h2[@class='ui-search-item__title shops__item-title')]
nombre_maquina = soup.find_all('h2',attrs={'class':'ui-search-item__title shops__item-title'})
```

```python
nombre_maquina

len(nombre_maquina)
```

##### <span style="color:blue">Pero como unicamente querermos todos títulos de cada máquina utilizaremos un ciclo (for).</span>

```python
nombre_maquina = [i.text for i in nombre_maquina]

nombre_maquina
```
