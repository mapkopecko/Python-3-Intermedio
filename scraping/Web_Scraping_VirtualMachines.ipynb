<h5 style="color:magenta">Vamos a utilizar (re) por que vamos a utilizar expresiones regulares</h5>

<h5 style="color:magenta">Vamos a utilizar (colorama) por que vamos a utilizar colores</h5>

<h5 style="color:magenta">Vamos a utilizar (request) para realizar peticiones HTTP a las Páginas Web</h5>

```python
import requests
```

<h5 style="color:magenta">Vamos a crear una variable llamada (website) para guardarle el link de la Página Web</h5>

```python
website = 'https://www.vulnhub.com/'
```

<h5 style="color:magenta">Vamos a utilizar la libreria (requests) para realizar la petición de la Página Web</h5>

```python
resultado = requests.get(website)
```

<h5 style="color:magenta">Vamos a convertir lo que asignamos a la variable (resultado) a texto de la Página Web</h5>

```python
content = resultado.text
```

<h5 style="color:magenta">Vamos a desplegamos en pantalla la petición (content) de la Página Web</h5>

```python
print(content)
```

<h5 style="color:magenta">Ahora vamos a buscar el título de cada máquina de la Página Web</h5>

<h5 style="color:magenta">Vemos que cada nombre de la máquina tiene un contexto y se ve un patrón en común en cada nombre de la máquina de la Página Web</h5>

<h5 style="color:magenta">Ahora le crearemos una expresión regular para obtener el nombre la máquina de la Página Web</h5>

<h5 style="color:magenta">Empezamos creando un (patroón) utilizando la libreria (r) para eliminar todas las palabras (/entry/) que encontremos en todas las lineas de los nombres de las máquinas y seguido de ([\w-]*) para obtener todo lo que esté despues de la palabra (/entry/) en la Página Web</h5>

<h5 style="color:magenta">Ahora para las máquinas repetidas creamos una variable llamada (maquinas_repetidas) que es igual a (re) que es la libreria para encontrar los nombred de las máquinas repetidas seguido de (findall) y entre parentesis el argumento (patron) y una (,) seguido de (content) que es la variable donde está almacenada nuestra Página Web y como esto funciona con strings debemos comvertirlo con (str). Con esto lo tenemos filtrado. Y lo desplegamos en pantalla.</h5>

```python
patron = r"/entry/[\w-]*"

maquinas_repetidas = re.findall(patron, str(content))

```

<h5 style="color:magenta">Como vemos en pantalla pone todos los nombres de las máquinas repetidas despues de la palabra (/entry/). Ahora vamos acomodar nuestro erreglo.</h5>

```python
print(maquinas_repetidas)
```

<h5 style="color:magenta">Ahora vamos a eliminar todos los elementos repetidos.</h5>

<h5 style="color:magenta">Primero crearemos una variable llamada (sin_duplicados) lo vamos a igular a una lista (list) y quitar los repetidos.</h5>

```python
sin_duplicados = list(set(maquinas_repetidas))
```

<h5 style="color:magenta">Lo mandamos a desplegar en pantalla.</h5>

```python
print(sin_duplicados)
```

<h5 style="color:magenta">Ahora vamos a desplegar cada nombre de esas máquinas pero el puro nombre y los pondremos en una lista. Utilizaremos un bucle (for)</h5>
<h5 style="color:magenta">Dentro del ciclo de iteración de nuestro listado de máquinas vamos a sustituir la palabra (/entry/) por un espacio vacio</h5>
<h5 style="color:magenta">Y el resultado lo vamos almacenar dentro de la lista (maquinas_final) </h5>

```python
maquinas_final = []

for i in sin_duplicados:
    nombre_de_maquina = i.replace("/entry/","")
    maquinas_final.append(nombre_de_maquina)
    print(nombre_de_maquina)
```

<h5 style="color:magenta">Vamos a crear un script donde me diga si han puesto una máquina nueva o hay un cambio en la página web</h5>
<h5 style="color:magenta">Vamos a crear una variable que se llame (maquina_noob) que me guarde el título de una máquina en la página web</h5>

```python
maquina_noob = "noob-1"
existe_noob = False
```

<h5 style="color:magenta">Vamos a crear bucle (for) para evaluar máquina por máquina si encuentra la máquina (noob-1) y si la encuantra o no y envie un mensaje en la página web</h5>
<h5 style="color:magenta">En nuestro bucle (for) utilizaremos nuestra lista creada donde se encuentra todas llamada (maquinas_final) y recorreremos todos los elementos de la lista en la página web</h5>

```python
for a in maquinas_final:
    if a == maquina_noob:
        existe_noob = True
        break


color_verde = Fore.GREEN

color_amarillo = Fore.YELLOW


if existe_noob == True:
    print("\n" + color_verde + "No hay ninguna máquina nueva...")
else:
    print("\n" + color_amarillo + "Existe una máquina nueva!!!!")
```
