#### <span style="color:yellow">2.- Enviar la petición HTTP para extraer el código fuente de la página WEB (página de prueba)</span>

#### <span style="color:yellow">La petición a la HTTP se envia por medio del método (requests) y se manda a desplegar en pantalla. y nos responde 200 nuestra petición fue aceptada de lo contrario mandará 404 que no fue aceptada.</span>

#### <span style="color:yellow">Una vez aceptada nuestra petición. Vamos a crear un objeto (soup) el cual toma como objeto (BeautifulSoup) y le mandamos nuestro response con lo que alla llegado de nuestra respuesta pagina web y accedemos a nuestro attributo texto. Y definimos un analizador o un parse en este caso.</span>

#### <span style="color:yellow">Una vez obtenido nuestra página web y nuestro objeto (soup) ya accedimos al código fuente de la página. Y la desplegamos con print().</span>

```python
pagina_web : requests.get('https://en.wikipedia.org/wiki/2014_FIFA_World_Cup')

soup = BeautifulSoup(pagina_web.text, 'html.parser')

print(soup.prettify())
```

#### <span style="color:yellow">Despues de esto vamos a crear un arreglo de los Años de las Copas Mundiales de Futball.</span>

#### <span style="color:yellow">Vamos a utilizar el método (find(Encuentra 1 elemento)) ó (find_all(Encuentra todos los elementos)) paraa inspeccionar la página web de las Copas Mundiales de Futball.</span>

```python
partidos_de_equipos = soup.find_all('div', class_='footballbox')
```

#### <span style="color:yellow">Vamos a crear un ciclo (for) para sacar todos los partidos o partido por partodo de la página web de las Copas Mundiales de Futball.</span>

```python
for partidos in partidos_de_equipos:
    partidos.find('span', intemprop='name').get_text()
    
```
