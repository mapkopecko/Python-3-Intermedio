##### <span style="color:yellow">Una vez aceptada nuestra petición podemos crear un objeto (soup) el cual toma como objeto BeautifulSoup y le mandamos la respuesta obtenida del requests y accedemos al atributo text y definimos un analizador en este caso html. ##### <span style="color:yellow">Lo colocas en tu variable creada (pagina_web) igualandolo. Después creas tu objeto (BeautifulSoup) igualado en tu variable (soup) donde vas a pasarle el texto de tu respuesta recibida de la variable (pagina_web) y lo despliegas en pantalla.</span>

##### <span style="color:yellow">La ventaja de ver la página web en su entorno que distinguir todas esas ramificaciones, todos estos nodos que se van desprendiendo. Desde la etiqueta HTML, la etiqueta HEAD, BODY y dentro de cada una de ellas lo que se va desprendiedo. Obeservamos que en nuetra etiqueta BODY es donde se desprende todos nuestros elementos o la información que son útiles para obtener la información deseada ó buscada. .</span>

```python
pagina_web = requests.get('https://www.cyberpuerta.mx/Computadoras/Tabletas/Tabletas/') # Un request con su url

print("Si despliega <Response [200]> (entonces tu petición fue escuchada y aceptada)  : ",pagina_web) # Se despliega el resultado obtenido por la petición
print()

soup = BeautifulSoup(pagina_web.text,'html.parser')

print(soup.prettify()) 
```
